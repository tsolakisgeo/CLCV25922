```python
# I spent a lot of time trying to make a solution work using a lambda formula, but that wasn't getting me anywhere
# (I kept getting an error on the "axis" in the df.apply() part). Eventually, I decided to start with two empty lists
# and add all the converted coordinates to them, before adding them to the new DataFrame as columns.

X = []
Y = []

for ind in UTM_df.index:
    Easting = utm.from_latlon(UTM_df['Latitude'][ind], UTM_df['Longitude'][ind])[0]
    Northing = utm.from_latlon(UTM_df['Latitude'][ind], UTM_df['Longitude'][ind])[1]
    
# This part of the loop iterates over each row in the DataFrame and calculates the UTM coordinates from the Longitude
# and Latitude of that row. The resulting tuple contains both Easting and Northing coordinates, but I specified the
# index of that tuple so only the correct coordinate would be matched.

    X.append(Easting)
    Y.append(Northing)

# I then appended those values to my empty lists, and repeated that for each index in the DataFrame.

UTM_df["Easting"] = X
UTM_df["Northing"] = Y

# Finally, I created two new columns in my DataFrame from the now-filled lists X and Y.
```
